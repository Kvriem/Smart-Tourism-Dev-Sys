## getting the most important tokens and delete the rubbish semantic tokens

```python
main_positive_tokens = {
    'staff', 'location', 'hotel', 'food', 'beach', 'nile', 'clean', 'room',
    'breakfast', 'view', 'service', 'pool', 'restaurant', 'reception', 'sea'
}

def extract_unique_matching_tokens(phrases):
    matched_tokens = set()
    for phrase in phrases:
        tokens = str(phrase).lower().split()
        matches = {token for token in tokens if token in main_positive_tokens}
        matched_tokens.update(matches)
    return list(matched_tokens)



df2['positive tokens'] = df2['Semantic Phrases Pos'].apply(extract_unique_matching_tokens)
```

```python
main_negative_tokens = ['room', 'hotel', 'food', 'wifi','beach', 'breakfast', 'noise','harrsing' , 'car parking'
                           'bathroom',  'staff', 'pool', 'bar','old','taxi', 'road','dirty','air conditioning',
                            'restaurant', 'clean', 'bed','price', 'service', 'location','sea','security',
                            'shower','bug','unproffessional','expensive','drinks','temperature']

def extract_unique_matching_tokens(phrases):
    matched_tokens = set()
    for phrase in phrases:
        tokens = str(phrase).lower().split()
        matches = {token for token in tokens if token in main_negative_tokens}
        matched_tokens.update(matches)
    return list(matched_tokens)

df2['negative tokens'] = df2['Semantic Phrases Neg'].apply(extract_unique_matching_tokens)
```
